```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
fig_width = 10
fig_height = 6
```

### What is the data bunch?

```python
# load data
iris_data_bunch = load_iris(as_frame=True)
print(type(iris_data_bunch))
print(iris_data_bunch.keys())
```

```python
print('data:')
print(iris_data_bunch['data'].head())
print()
print(iris_data_bunch['data'].shape)
```

```python
print('target:')
print(iris_data_bunch['target'].head())
```

```python
print('frame:')
print(iris_data_bunch['frame'].head())
```

```python
print('target names:')
print(iris_data_bunch['target_names'])
```

```python
print('DESCR:')
print(iris_data_bunch['DESCR'])
```

```python
print('feature names:')
print(iris_data_bunch['feature_names'])
```

```python
print('filename:')
print(iris_data_bunch['filename'])
```

### Visualize the data

```python
fig, ax = plt.subplots()
tf = (iris_data_bunch['target']==0)
ax.scatter(iris_data_bunch['data'].loc[tf, 'sepal length (cm)'], iris_data_bunch['data'].loc[tf, 'sepal width (cm)'], \
           color='red', label='setosa')

tf = (iris_data_bunch['target']==1)
ax.scatter(iris_data_bunch['data'].loc[tf, 'sepal length (cm)'], iris_data_bunch['data'].loc[tf, 'sepal width (cm)'], \
           color='blue', label='versicolor')

tf = (iris_data_bunch['target']==2)
ax.scatter(iris_data_bunch['data'].loc[tf, 'sepal length (cm)'], iris_data_bunch['data'].loc[tf, 'sepal width (cm)'], \
           color='green', label='verginica')

ax.legend()
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
```
