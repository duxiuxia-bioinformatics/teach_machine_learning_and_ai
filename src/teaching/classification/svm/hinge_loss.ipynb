```python
import numpy as np
import matplotlib.pyplot as plt

# Define range of y*f(x)
z = np.linspace(-2, 3, 400)

# Hinge loss: max(0, 1 - z)
hinge_loss = np.maximum(0, 1 - z)

# Plot
plt.figure(figsize=(6, 4))
plt.plot(z, hinge_loss, color='red')

# Add reference lines
# plt.axvline(x=0, linestyle='--')   # decision boundary
# plt.axvline(x=1, linestyle='--')   # margin boundary
# plt.axhline(y=0, linestyle='--')

# Labels
plt.xlabel(r'$y(w^T x + b)$')
plt.ylabel("Hinge Loss")
plt.title("Hinge Loss Function")

plt.grid(True)

out_file_name = 'hinge_loss.png'
plt.savefig(out_file_name)

plt.show()
```

```python
logistic_loss_positive = np.log(1 + np.exp(-z))
logistic_loss_negative = np.log(1 + np.exp(z))

plt.figure(figsize=(6,4))
plt.plot(z, hinge_loss, color = 'red', label="Hinge Loss")
plt.plot(z, logistic_loss_positive, linestyle='--', label="Logistic Loss for y = 1")
plt.plot(z, logistic_loss_negative, linestyle='--', label="Logistic Loss for y = -1")

plt.xlabel(r'$y(w^T x + b)$')
plt.ylabel(r'$J=\log (1 + e^{-y \cdot z})$')
plt.title("Hinge vs Logistic Loss")
plt.legend()
plt.grid(True)

out_file_name = 'hinge loss vs logistic loss.png'
plt.savefig(out_file_name)

plt.show()
```

The logistic loss is defined as:
$$
J = \log(1 + e^{-t \cdot z})
$$
where $t = 1$ or $t = -1$. This logistic loss is the same as the binary cross-entropy loss where labels are $y=0$ or $y=1$:
$$
\text{binary cross-entropy cost} = - \left[ y \log p + (1-y) \log (1-p) \right]
$$
where
$$
p = \frac{1}{1+e^{-z}}, \quad z = w^T x + b
$$
The logistic loss is smooth and probabilistic, while hinge loss is margin-based and becomes exactly zero once the margin is satisfied.
