Goal:
- To compute Gini Impurity and information gain by hand
- To verify your hand-calculated Gini Impurity and information gain using python


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data
in_file_name = '../../data/data_for_entropy_and_information_gain.csv'
df = pd.read_csv(in_file_name)
print(df)
```

```python
def gini(y):
    y = np.array(y)
    n = len(y)
    if n == 0:
        return 0.0
    vals, counts = np.unique(y, return_counts=True)

    ps = counts / n
    gini = float(1 - np.sum(ps**2))
    return gini

def gini_info_gain(df, feature, target="Disease (Y)"):
    H = gini(df[target])
    n = len(df)

    cond = 0.0
    for val, sub in df.groupby(feature):
        print(f'val = {val}')
        print(type(sub))
        print(f'sub = {sub}\n')
        cond += (len(sub)/n)*gini(sub[target])

    IG = H - cond
    return IG

H_all = gini(df["Disease (Y)"])
IG_S = gini_info_gain(df, "Smoking (x1)")
IG_M = gini_info_gain(df, "Mutation (x2)")

H_all, IG_S, IG_M
```

**Expected outputs (rounded):**
- Gini Impurity of full dataset: ~0.48 bits
- IG(Smoking): ~0.08 bits
- IG(Mutation): ~0.36 bits  
So **Mutation** is the best root split (higher information gain).
