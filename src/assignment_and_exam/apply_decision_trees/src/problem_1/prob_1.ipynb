Goal:
- To compute entropy and information gain by hand
- To verify your hand-calculated entropy and information gain using python

```python
import numpy as np
import pandas as pd
```

```python
# import data
in_file_name = '../../data/data_for_entropy_and_information_gain.csv'
df = pd.read_csv(in_file_name)
df
```

```python
def entropy(y):
    y = np.array(y)
    n = len(y)
    if n == 0:
        return 0.0

    vals, counts = np.unique(y, return_counts=True)
    ps = counts / n
    entropy = float(-np.sum([p*np.log2(p) for p in ps if p > 0]))
    return entropy

def entropy_info_gain(df, feature, target="Disease (Y)"):
    H = entropy(df[target])
    n = len(df)
    cond = 0.0
    for val, sub in df.groupby(feature):
        cond += (len(sub)/n)*entropy(sub[target])
    return H - cond

H_all = entropy(df["Disease (Y)"])
IG_S = entropy_info_gain(df, "Smoking (x1)")
IG_M = entropy_info_gain(df, "Mutation (x2)")

H_all, IG_S, IG_M
```

**Expected outputs (rounded):**
- Entropy of full dataset: ~0.971 bits
- IG(Smoking): ~0.124 bits
- IG(Mutation): ~0.610 bits

So **Mutation** is the best root split (higher information gain).
