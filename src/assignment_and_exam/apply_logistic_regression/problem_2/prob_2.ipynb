Goal:
- To manually compute log-odds, odds, probability, and odds-ratio



Model:  $$\log \left( \frac{p}{1-p} \right) = -2 + 0.8\cdot \text{GeneA} $$

```python
import math

def sigmoid(z):
    return 1/(1+math.exp(-z))

# (a) GeneA=0
z0 = -2 + 0.8*0
p0 = sigmoid(z0)

# (b) GeneA=3
z3 = -2 + 0.8*3
p3 = sigmoid(z3)

# (c) OR for one-unit increase
OR = math.exp(0.8)

p0, p3, OR
```
