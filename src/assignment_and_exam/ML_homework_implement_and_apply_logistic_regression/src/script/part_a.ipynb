```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

#### (a) (10 points) Do natural log transform of the PFOS variable in file $\texttt{pfas.csv}$ and store the results as a new variable log\_PFOS in the data file. Standardize the variables $x$=[log\_PFOS, age, gender, BMI].

```python
# import the data
in_file_name = '../../data/pfas.csv'
data_in_df = pd.read_csv(in_file_name)
data_in_df.head()
```

```python
# Take the natural log transform of the PFOS variable
data_in_df['log_PFOS'] = np.log(data_in_df['PFOS']) 
data_in_df.head()
```

```python
# Standardize all the variables
data_for_standardization_df = data_in_df.drop(columns=['disease'])
standard_scaler_obj = StandardScaler()
standard_scaler_obj.fit(data_for_standardization_df)
data_after_standardization_array = StandardScaler().fit_transform(data_for_standardization_df)
data_after_standardization_df = pd.DataFrame(data_after_standardization_array, \
                                             columns=data_for_standardization_df.columns, \
                                             index=data_for_standardization_df.index)
data_standardized_df = pd.concat([data_in_df[['disease']], data_after_standardization_df], axis=1)
```

```python
# Export the standardized data

out_file_name = 'pfas_standardized.csv'
data_standardized_df.to_csv(out_file_name, index=False)
```
