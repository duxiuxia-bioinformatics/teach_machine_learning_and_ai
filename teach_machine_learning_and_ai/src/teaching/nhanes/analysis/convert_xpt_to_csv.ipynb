```python
import pandas as pd
import os
```

```python
in_file_path = '../data/original'
in_file_name = 'PFAS_J.xpt'
in_file_full_name = os.path.join(in_file_path, in_file_name)
data_df = pd.read_sas(in_file_full_name, format='xport')
print(data_df.shape)
```

```python
# export the data into csv file
out_file_path = '../data/converted/'
ind = in_file_name.find('.xpt')
file_identifier = in_file_name[:ind]

out_file_name = file_identifier + '.csv'
out_file_full_name = os.path.join(out_file_path, out_file_name)
print(out_file_name)
data_df.to_csv(out_file_full_name, index=False)
```
